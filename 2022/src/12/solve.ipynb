```python
#bfs #graphs

f = open('input.txt', 'r')
rows = f.read().split('\n')

def find2d(rows, letter):
    for i in range(len(rows)):
        res = rows[i].find(letter)
        if res != -1:
            return f'{i}.{res}'


def build_graph(vertices):
    graph_ = dict()
    for i in range(len(vertices)):
        for j in range(len(vertices[i])):
            graph_[f'{i}.{j}'] = list()
            
            cur = ord(vertices[i][j])
            up = ord(vertices[:i][::-1][0:1] [0][j] if vertices[:i][::-1][0:1] else 'ีก')
            down = ord(vertices[i+1:][0:1] [0][j] if vertices[i+1:][0:1] else 'ีก')
            right = ord(vertices[i][j+1:][0:1][0] if vertices[i][j+1:][0:1] else 'ีก')
            left = ord(vertices[i][:j][::-1][0:1][0] if vertices[i][:j][::-1][0:1] else 'ีก')
            
            if cur - up >= -1:
                graph_[f'{i}.{j}'].append(f'{i-1}.{j}')
            if cur - down >= -1:
                graph_[f'{i}.{j}'].append(f'{i+1}.{j}')
            if cur - right >= -1:
                graph_[f'{i}.{j}'].append(f'{i}.{j+1}')
            if cur - left >= -1:
                graph_[f'{i}.{j}'].append(f'{i}.{j-1}')
    return graph_


def reverse_graph(graph):
    reversed_graph = dict()
    for vertex in graph:
        for v in graph.get(vertex):
            if reversed_graph.get(v) is None:
                reversed_graph[v] = list()
            reversed_graph[v].append(vertex)
    return reversed_graph

    
def bfs(graph, start, end, vertices=None):    
    
    queue = []
    queue.append([start])
    
    in_search = []
    in_search.append(start)

    while queue:
        path = queue.pop(0)
        node = path[-1]
        
        # for part II
        if vertices:
            i, j = [int(x) for x in node.split('.')]
            node_ = vertices[i][j]
        else:
            node_ = node
            
        if node_ == end:
            return path

        for adjacent in graph.get(node, []):
            if adjacent not in in_search:
                new_path = list(path)
                new_path.append(adjacent)
                queue.append(new_path)                
                in_search.append(adjacent)
    return None

                
vertices = list()

for str in rows:
    str = str.replace('S', 'a').replace('E', 'z')
    vertices.append(list(map(lambda x: (x), str)))              
                

graph_ = build_graph(vertices)        

init_point = find2d(rows, 'S')
dest_point = find2d(rows, 'E')

# part I
path = bfs(graph_, init_point, dest_point)
print(path)
print(len(path) - 1)


# part II
r_graph_ = reverse_graph(graph_)
path = bfs(r_graph_, dest_point, 'a', vertices)
print(path)
print(len(path) - 1)
```
